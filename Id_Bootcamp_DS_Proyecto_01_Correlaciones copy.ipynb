## Analizar precio en función a numero de habitaciones, para propiedades en venta

```python
data = data.loc[(data['operation_type']=='Venta') & (data['rooms'].notnull()) & (data['price'].notnull()) & (data['rooms']<16)]
```

```python
data = data.loc[data['price']<2_000_000]
```

```python
data['l2'].unique()
```

```python
data['property_type'].unique()
```

```python
data_a_dibujar = data.loc[data['property_type']=='Departamento']
sns.scatterplot(data=data_a_dibujar, x="rooms", y="price")
plt.show()
```

```python
def dibujar_relacion(df, tipo_propiedad):
    data_a_dibujar = data.loc[data['property_type']==tipo_propiedad]
    sns.scatterplot(data=data_a_dibujar, x="rooms", y="price").set(title = f"Tipo de propiedad {tipo_propiedad}")
    plt.show()
```

```python
dibujar_relacion(df=data, tipo_propiedad='Oficina')
```

```python
dibujar_relacion(df=data, tipo_propiedad='Departamento')
```

```python
def dibujar_relacion_por_ciudad(df, tipo_propiedad, ciudad):
    data_a_dibujar = data.loc[(data['property_type']==tipo_propiedad) & (data['l2']==ciudad)]
    sns.scatterplot(data=data_a_dibujar, x="rooms", y="price").set(title = f"Tipo de propiedad {tipo_propiedad}")
    plt.show()
```

```python
dibujar_relacion_por_ciudad(df=data, tipo_propiedad='Casa', ciudad='Mendoza')
```

```python
dibujar_relacion_por_ciudad(df=data, tipo_propiedad='Departamento', ciudad='San Juan')
```
